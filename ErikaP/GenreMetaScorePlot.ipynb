```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
GENRE_CSV_PATH = "GenreMetaScore.csv"
IMDB_YELLOW = '#F5DE50'
```

```python
genre_df = pd.read_csv(GENRE_CSV_PATH) 
print(genre_df.shape)
genre_df.head(3)
```

```python
# Learning about series
print(genre_df['Genre'])
print(type(genre_df['Genre']))
print(genre_df['Genre'].index)
print(genre_df['Genre'].values)    # x-values
```

```python
genre_df['Meta_Score'].values
```

```python
#container = plt.bar(genre_df['Genre'].values, genre_df['Meta_Score'].values, color=IMDB_YELLOW)
# plt.bar_label(container, labels=genre_df['Top_Director'].values, label_type='center')

genres_directors = (genre_df['Genre'] + '\n' + genre_df['Top_Director']).values
plt.bar(genres_directors, genre_df['Meta_Score'].values, color=IMDB_YELLOW)
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Meta Score')
plt.title('Meta Score vs. Genre')

plt.show()
```
