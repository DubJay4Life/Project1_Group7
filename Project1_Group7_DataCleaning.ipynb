```python
#Column names are good
#Drop poster, overview
#Null values - 3 columns - drop becuase if gross is 0 or null it won't help 
#Run time, gross, genre clean
```

```python
df2 = df.dropna().reset_index(drop=True)
df2.info()
```

```python
# Split and extract the number
df2['Runtime'] = df2['Runtime'].str.split().str[0].astype(int)
df2.info()
```

```python
# Remove commas and convert to integers
df2['Gross'] = df2['Gross'].str.replace(',', '').astype(int)
df2.info()
```

```python
# Count the number of genres
df2['Num_Genres'] = df2['Genre'].str.split(',').str.len()

# Extract the first genre using split
df2['Genre'] = df2['Genre'].str.split(',').str[0]
df2.info()
```

```python
df2.head()
```

```python
df2.Certificate.value_counts()
```

```python
#Limitation
df2["Certificate"] = df2.Certificate.replace({
    "U": "G",
    "A": "R",
    "UA": "PG-13",
    "U/A": "PG-13",
    "TV-PG": "PG",
    "GP": "PG",
    "Approved": "PG",
    "Passed": "PG",
})
df2.Certificate.value_counts()
```

```python
df2 = df2.drop(columns=["Poster_Link", "Overview", "Star2", "Star3", "Star4"])
df2.head()
```

```python
df2 = df2.rename(columns={
    "Series_Title": "Movie_Title",
    "Certificate": "Rating",
})
df2.head()
```

```python
df2.to_csv("MovieData_Clean.csv", index=False)
```
